```python
import json
import os
import pandas as pd
```

```python
root_dir = "../results/uci_har_tinyhar_ep20_pat5_bs128_lr0.0003"

# message if the directory does not exist
if os.path.exists(root_dir) is False:
    raise FileNotFoundError(f"The directory {root_dir} does not exist.")
```

```python
all_metrics = []
for dir in os.listdir(root_dir):
    metrics_path = os.path.join(root_dir, dir, "metrics.json")

    # Safety check to ensure the file exists
    if os.path.exists(metrics_path):
        with open(metrics_path, "r") as f:
            metrics = json.load(f)

        # Flatten the nested structure
        flat_metrics = {}
        for split, values in metrics.items():
            for metric_name, value in values.items():
                flat_metrics[f"{split}_{metric_name}"] = value

        all_metrics.append(flat_metrics)

# Create DataFrame
df = pd.DataFrame(all_metrics)

# Standardize output
print("--- Mean Test Macro F1 ---")
print(df["test_f1_macro"].mean())
print("\n--- Median Test Macro F1 ---")
print(df["test_f1_macro"].median())
```
